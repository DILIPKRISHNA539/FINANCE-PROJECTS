```python
import pandas as pd
import yfinance as yf
from datetime import datetime

start_date = datetime.now() - pd.DateOffset(months=3)
end_date = datetime.now()

tickers = ['SONY', 'ADANIPORTS.NS', 'TSLA', 'NFLX', 'TATAMOTORS.NS']

df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data)

df = pd.concat(df_list, keys=tickers, names=['Ticker', 'Date'])
print(df.head())
```

```python
df
```

```python
df=df.reset_index()
print(df.head())
```

```python
import plotly.express as px
fig= px.line(df, x='Date', y='Close', color='Ticker', title='Stock Price Over 3 months')
fig.show()
```

```python
fig = px.area(df, x='Date', y='Close', color='Ticker', facet_col='Ticker',
labels = {'Date': 'Date', 'Close': 'Closing Price', 'Ticker': 'Company'},
title='Stock Prices of SONY, ADANIPORTS, TESLA, NETFLIX AND TATAMOTORS')
fig.show()
```

```python
df['MA10']= df.groupby('Ticker')['Close'].rolling(window=10).mean().reset_index(0,drop=True)
df['MA20']= df.groupby('Ticker')['Close'].rolling(window=10).mean().reset_index(0,drop=True)

for ticker, group in df.groupby('Ticker'):
    print(f'Moving Averages for {ticker}')
    print(group[['MA10', 'MA20']])
```

```python
for ticker, group in df.groupby('Ticker'):
    fig = px.line(group, x='Date', y=['Close', 'MA10', 'MA20'],
                  title=f"{ticker} Moving Averages")
    fig.show()
```

```python
df['Volatility'] = df.groupby('Ticker')['Close'].pct_change().rolling(window=10).std().reset_index(0, drop=True)
fig = px.line(df, x='Date', y='Volatility',
              color='Ticker',
              title='Volatility of All Companies')
fig.show()
```

```python
tesla = df.loc[df['Ticker'] == 'TSLA', ['Date', 'Close']].rename(columns={'Close': 'TSLA'})
adaniports = df.loc[df['Ticker'] == 'ADANIPORTS.NS', ['Date', 'Close']].rename(columns={'Close': 'ADANIPORTS.NS'})
df_corr = pd.merge(tesla, adaniports, on='Date')

# create a scatter plot to visualize the correlation
fig = px.scatter(df_corr, x='TSLA', y='ADANIPORTS.NS',
                 trendline='ols',
                 title='Correlation between TESLA and ADANIPORTS.NS')
fig.show()
```
